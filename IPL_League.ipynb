```python
# a) Importing pandas and python library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# b) Loading Data

matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

matches.head(10)
```

```python
# 1. Matches won by the maximum margin of runs

max_runs = matches.loc[matches['win_by_runs'].idxmax()]
max_runs[['id', 'season', 'city', 'date', 'team1', 'team2', 'winner', 'win_by_runs', 'venue']]
```

```python
# 2. Matches won by maximum wickets

max_wickets = matches.loc[matches['win_by_wickets'].idxmax()]
max_wickets[['id', 'season', 'city', 'date', 'team1', 'team2', 'winner', 'win_by_wickets', 'venue']]
```

```python
# 3. Matches won by the minimum margin of runs

min_runs = matches[matches['win_by_runs'] > 0].sort_values('win_by_runs').iloc[0]
min_runs[['id', 'season', 'city', 'date', 'team1', 'team2', 'winner', 'win_by_runs', 'venue']]
```

```python
# 4. Matches won by minimum wickets

min_wickets = matches[matches['win_by_wickets'] > 0].sort_values('win_by_wickets').iloc[0]
min_wickets[['id', 'season', 'city', 'date', 'team1', 'team2', 'winner', 'win_by_wickets', 'venue']]
```

```python
# 5. Matches where D/L method was and wasn't applied

matches['dl_applied'].replace({ 0 : "Not Applied", 1 : "Applied"}).value_counts()
```

```python
# 6. Numbers of Matches held in each city

city_matches = matches['city'].value_counts().head(10)

city_matches.plot(kind= 'bar')
plt.title("Top Cities Hosting IPL Matches")
plt.xlabel("City")
plt.ylabel("Number of Matches")
plt.xticks(rotation = 45)
plt.show()
```

```python
# 7. Number of Matches won by each team

team_wins = matches['winner'].value_counts().head(10)

team_wins.plot(kind = 'bar')
plt.title("Top Teams won the Matches")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation = 90)
plt.show()
```

```python
# 8. Number of Matches held every season

season_matches = matches['season'].value_counts().sort_index()

season_matches.plot( kind= 'line', marker = '.')
plt.title("Matches Played per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()
```

```python
# 9. Top 10 players based on No. of Man Of the Match awards won

matches['player_of_match'].value_counts().head(10)
```

```python
# 10. Does winning the Toss mean winning the match?

toss_matches_wins = matches[matches['toss_winner'] == matches['winner']]
len(toss_matches_wins) / len(matches) * 100
```

```python
# 11. What was the decision taken by captains when they won the toss?

matches['toss_decision'].value_counts()
```

```python
# 12. Number of Matches where D/L method was applied every season

matches[matches['dl_applied'] == 1]['season'].value_counts().sort_index()
```

```python
# 13. Different results of Matches

matches['result'].value_counts()
```

```python
# 14. How many times did a team win the toss?

matches['toss_winner'].value_counts()
```

```python
# b) Chasing Venue

matches[matches['win_by_wickets'] > 0]['venue'].value_counts().head(10)
```

```python
# b) Chasing Team

matches[matches['win_by_wickets'] > 0]['winner'].value_counts().head(10)
```
